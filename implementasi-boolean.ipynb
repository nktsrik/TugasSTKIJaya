```python
import re
import nltk
from nltk.stem import PorterStemmer
```

```python
bop = {
    'tidak': 3,
    'dan': 2,
    'atau': 1,
    '(': 0,
    ')': 0
}

BO = list(bop.keys())
BO
```

```python
masuk = ""
masukan = input(f"masukan kata:{masuk}")
```

```python
list_of_masukan = masukan.split()
list_of_masukan
```

```python
st = PorterStemmer()
```

```python
clean_input = []
for index, teks in enumerate(list_of_masukan):
    clean = re.sub(r'[^\w\s]','',teks.lower())
    stemword = st.stem(clean)
    clean_input.append(stemword)
print(clean_input, sep=', ')
```

```python
BOP = {
    'tidak': 3,
    'dan': 2,
    'atau': 1,
    '(': 0,
    ')': 0
}

infix = clean_input  # Anda perlu mendefinisikan variabel clean_input sebelumnya
postfix = []
stack = []

# Fungsi untuk mendapatkan prioritas operator
def precedence(operator):
    return BOP.get(operator, -1)

# Iterasi melalui setiap karakter dalam infix
for token in infix:
    # Jika token bukan operator, tambahkan ke postfix
    if token not in BOP:
        postfix.append(token)
    else:
        # Jika token adalah '(', langsung tambahkan ke stack
        if token == '(':
            stack.append(token)
        # Jika token adalah ')', pop operator dari stack ke postfix
        elif token == ')':
            top = stack.pop()
            while top != '(':
                postfix.append(top)
                top = stack.pop()
        # Jika token adalah operator
        else:
            # Pop operator-operator dengan prioritas lebih tinggi dari token
            while stack and precedence(stack[-1]) >= precedence(token):
                postfix.append(stack.pop())
            stack.append(token)

# Pop semua operator yang tersisa dari stack ke postfix
while stack:
    postfix.append(stack.pop())

# postfix sekarang berisi ekspresi dalam bentuk postfix
```

```python
postfix
```

```python
mappedQuery = {}
BO = ['dan', 'atau', 'tidak']
for index, var in enumerate(bag_of_words):
    templist = []
    for item in postfix:
        if item in BO:
            templist.append(item)
        else:
            if item in var:
                templist.append(True)
            else:
                templist.append(False)
    mappedQuery[index] = templist
```

```python
print(mappedQuery, sep=', ')
```
