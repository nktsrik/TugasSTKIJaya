```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Membaca data dari file CSV
dataset_id = pd.read_csv("clean-Data-Corpus-IND.csv", index_col=0)

# Menghapus baris yang mengandung nilai NaN
dataset_id.dropna(inplace=True)

# Mengambil kolom teks dan mengonversi menjadi list
corpus_id = dataset_id["teks"].tolist()

# Inisialisasi model TF-IDF
tr_idf_model_id = TfidfVectorizer()

# Melakukan fit transform pada corpus untuk mendapatkan vektor TF-IDF
tf_idf_vector_id = tr_idf_model_id.fit_transform(corpus_id)

# Sekarang Anda dapat menggunakan tf_idf_vector_id untuk analisis lanjutan
```

```python
print(type(tf_idf_vector_id), tf_idf_vector_id.shape)
```

```python
tf_idf_array_id = tf_idf_vector_id.toarray()

print(tf_idf_array_id)
```

```python
words_set_id = tr_idf_model_id.get_feature_names_out()

print(words_set_id)
```

```python
df_tf_idf_id = pd.DataFrame(tf_idf_array_id, columns = list(words_set_id))

df_tf_idf_id
```

```python
result_id = pd.concat([df_tf_idf_id, validate_id.teks], axis=1)
result_id.head(2)
```
